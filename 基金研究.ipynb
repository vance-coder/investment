## 计算横盘情况

```python
from itertools import groupby

def my_round(num, by=0.05):
    tmp = num / by
    return round(round(tmp, 0) * by, 2)
df['new'] = df['累计净值'].map(my_round)

by = 0.05
counts = 40
lst = df['new'].tolist()
dt = df['净值日期'].tolist()
vals = df['累计净值'].tolist()
data = [0,]*len(dt)
for point, cot in df['new'].value_counts().to_dict().items():
    if cot < counts:
        # 出现次数少可以直接pass
        continue
    
    tmp_lst = [point if ((point + by == val) or (point - by == val) or (point == val)) else 0 for val in lst]
    tmp_idx = 0
    for i,j in groupby(tmp_lst):
        _len = len(list(j))
        if i != 0 and _len >= counts:
            # print(i, _len, dt[tmp_idx], dt[tmp_idx+_len-1])
            
            # 计算值
            tmp_val = round(sum(vals[tmp_idx:tmp_idx+_len-1])/_len, 3)
            data[tmp_idx:tmp_idx+_len-1] = [tmp_val,]*_len
        tmp_idx += _len
    
# 合并区间   
```

```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts import options as opts
from pyecharts.charts import Kline, Line, Grid
from IPython.display import display

c = (
    Line()
    .add_xaxis(df['净值日期'].tolist())
    .add_yaxis(
        "累计净值", 
        df['累计净值'].tolist(),
        # markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="average", name='均值')]),
        is_hover_animation=False,
        linestyle_opts=opts.LineStyleOpts(width=1.5, opacity=0.8),
        label_opts=opts.LabelOpts(is_show=False)
        # markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(y=df['new'].tolist(), name='均值')])
    )
    .add_yaxis(
        "分段均值", 
        data,
        is_smooth=False,
        is_hover_animation=False,
        linestyle_opts=opts.LineStyleOpts(width=1.5, opacity=0.8, type_='dotted'),
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(
        yaxis_opts=opts.AxisOpts(
            is_scale=True,
            splitarea_opts=opts.SplitAreaOpts(
                is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
            ),
        ),
        xaxis_opts=opts.AxisOpts(is_scale=True),
        datazoom_opts=[opts.DataZoomOpts(type_="inside")],
        title_opts=opts.TitleOpts(title='008087'),
    )
    # .set_global_opts(title_opts=opts.TitleOpts(title="008087"))
)
# 最高下跌到目前点位

c.load_javascript()
```
